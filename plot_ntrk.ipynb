```python
import ROOT as rt
from ROOT import RDataFrame, TH1D
```

```python
df = RDataFrame("detector_tree", "out.root")
```

```python
df = df.Define("num_tracks", "tracks_npoints.size()")
```

```python
hx = rt.RDF.TH1DModel(TH1D("hx", "hx", 10, 0, 20))
```

```python
hx = df.Histo1D(hx, "nhits")
```

```python
hx = hx.GetValue()
```

```python
cx = rt.TCanvas()
hx.Draw()
cx.Draw()
```
