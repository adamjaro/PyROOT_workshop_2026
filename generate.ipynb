# Run to generate the events

initial imports:

```python
from ctypes import c_int, c_double
from ROOT import TFile, TTree, std, TF1, TF2
```

Output file and tree

```python
out = TFile("out.root", "recreate")
tree = TTree("detector_tree", "detector_tree")
```

Forward hits

```python
nhits = c_int(0)
tree.Branch("nhits", nhits, "nhits/I")
```

Track points

```python
tracks_npoints = std.vector("int")()
tree.Branch("tracks_npoints", tracks_npoints)
tracks_eta = std.vector("double")()
tree.Branch("tracks_eta", tracks_eta)
```

Landau distribution for hits

```python
landau = TF1("landau", "TMath::Landau(x,[0],[1])", 0, 40)
landau.SetParameters(4, 1)
```

Number of events to generate

```python
nev = 12
```

Event loop

```python
for iev in range(nev):
    nhits.value = int(landau.GetRandom())
    tree.Fill()
```

Write output tree and close

```python
tree.Write()
out.Close()
```
